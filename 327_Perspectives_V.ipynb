Use Google news searches or library database searches to find a news article about an LGBTQ current event from within the last 5 years. This assignment hones your researching and critical thinking skills, particularly around identifying positionality and intersectional analysis. #### What is Homonationalism?
- the term is sometimes used derisively to describe any campaign for LGBTQ rights that could be coopted by nationalist or militaristic agendas
- it is not a political position to be identified with, it is a structure through which we can analyze the ways in which LGBTQ rights are often framed and deployed
- in Terrorist Assemblages, Jasbir Puar argued against the notion that the state is solely heteronormative and excludes queer individuals at all times and in all ways
    - in Rethinking Homonationalism, Puar goes on to say that homonationalism is exemplified by how queer acceptance has become "a barometer by which the right to and capacity for national sovereignty is evaluated"
        - this means that in some way or other, all LGBTQ rights campaigns uphold some ideals of social progress and act on the assumption that western nation states are capable of accepting and benefiting all people from all groups
        - Puar argues that this is impossible as these states are built upon granting privileges to certain groups, necessitating that they be taken away from others
            - e.g. the fight for and granting of marriage equality legitimize and entrench the normative ideal of marriage between two people
            - e.g. a state granting trans citizens the right to change their identity on a passport legitimizes the state's ability to control and surveil its citizens, to control movement between borders, and to deport racialized minorities
- it is difficult not to recreate homonationalist discourses whenv discussing current events such as Trump's push to ban trans servicemembers from the military
    - i.e. it can be difficult to decouple support from trans people in the military from support for the military itself
        - "trans people are just as good at killing brown people overseas as cis people."

#### Sexual and Gender Diversity in Native America and the Pacific Islands
- multiple gender and sexual identities have existed in Native cultures on mainland North America and the Pacific Islands
    - european explorers often noted the prevalence of "hermaphrodites" and "sodomites" within these cultures
    - europeans lacked both context and language to understand these identities
    - the term "berdache" was used to describe identities which did not fit within the European binary
        - this term is now considered derogatory and is not used by Native people
        - in 1990, the term "Two-Spirit" was coined to describe identities balancing or combining male and female qualities
- two-spirit
    - prior to the arrival of Europeans, many Native cultures had their own terms and concepts for these identities
        - for some, two spiritedness was noted in childhood, some held ceremonies, and some staged ritual tests to determine gender identity
    - male two spirits often fell into intermediate gender roles, working hides, conducting burial rites, and serving as healers and ceremony leaders
        - some ceremonial functions were reserved for two spirits
    - there is less evidence for females engaged in men's activities
        - it is worth noting that Euro-American anthropology tends to overlook these sorts of identities
        - women also often engaged in male pursuits without being labeled as two-spirited
    - two spirits generally formed relationships with non-two-spirit individuals of the same sex
        - these relationships were seen as equally valid to their heterosexual counterparts
- little was recorded of Hawai'i's mahu, men who performed women's work and formed relationships with other men
    - oral traditions credit them with important roles in society including healing, caretaking, and leading hula dance traditions
    - aikane relationships were distinct from this and held sexual connotations
        - these tended to but were not required to be between men of different ages and/or social status
        - most men with aikane were bisexual and married women as well
- a recent revival
    - in the past 50 or so years, there has been a resurgence of interest in two-spirit, mahu, and similar identities
        - natives seek to reclaim their identities and traditions after centuries of erasure and forced assimilation
