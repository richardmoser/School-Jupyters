### Instructions and Rubric

**Instructions**
This assignment is designed to engage with the world that surrounds you from a slightly different perspective: one that is particularly focused upon LGBTQ activism. Battles over and about sexuality are happening all around us, all the time. Groups of sexual minorities are constantly fighting back against their persecution and discrimination. This activism can be transformational but also reinforce other forms of oppression (often without thought or intention).

Use Google news searches or library database searches to find a news article about an LGBTQ current event from within the last 5 years. Using the news article as a jumping off point, do a bit more investigating into the issue you’ve chosen. Then pick a group or form of protest that addressed the issue directly. Finally, you will create a 3 to 4-minute video (much like a newscast segment) that does the following:
- Describes the persecution and an activist response to the issue
- Analyzes how the activism challenges the initial oppression
- Deploys an intersectional analysis to interrogate whether the activism rearticulates other forms of oppression. If it does, explore how and in what ways it affirms other forms of oppression. If it does not, how do the activists involved accomplish their activism without relying on the oppression of others?

In addition to the video, you must also upload a bibliography of your news source and any other research you drew from in the making of your newscast.

This assignment hones your researching and critical thinking skills, particularly around identifying positionality and intersectional analysis. This assignment can also expand the ways you think about activism and spark ideas about how to change the world for the better. Intersectional activism is necessary but also HARD to do!


**Rubric**

**Identify and explain your chosen current event: 20%**
This is largely a summary of your chosen current event. However, you also need to provide some larger context about the time and place relating to your choice.

**Identify and explain an activist response or strategy: 20%**
Here, you should not just describe what activists do, but assess the strategy behind it. Why did they choose that kind of response? Were other options considered? Are you focusing on one group’s response that was coordinating with other groups to give a more dynamic response? How was the activism received? What changes occurred as a result?

**Intersectional Analysis: 25%**
Even well-intentioned activists make mistakes, have ways to grow, and can be oppressors of others. In fact, existing in the world without reinforcing and benefitting from/experiencing oppression of others is VERY HARD for all of us all the time. This is a chance to really explore how intentional, strategic, and unintentional oppression happens. Look for ways in which activists made assumptions about either other activists or those they were protesting. Was the activism accessible for people with different physical capabilities? Did a gay or lesbian group decide to leave transgender issues off their agenda? Was there an assumption about race or wealth? Once you identify these forms of oppression, go deeper and try to discern if this was done intentionally. If so, was it out of malice, ignorance, strategy? Was it done unintentionally? What can be learned from that?

**Newscast composition: 15%**
Just because this isn’t a written assignment does not mean that it needs to be any less structured or argument driven than your other assignments. Plan out how you will address all the requirements of the prompt. You might consider making an outline, or even a script. Speak clearly and slowly enough so that I can understand you, please! Remember that the video should be between 3 and 4 minutes long; points will be taken off if your video is shorter than 2.75 minutes or longer than 4.25 minutes.

**Forward Thinking: 10%**
Finally, make some suggestions to future activists about how they might avoid reinforcing other forms of oppression.

**Cite your sources: 10%**
Provide a full citation for each source you reference in your video and submit it as a Works Cited/Bibliography in addition to your video.

#### What is Homonationalism?
- the term is sometimes used derisively to describe any campaign for LGBTQ rights that could be coopted by nationalist or militaristic agendas
- it is not a political position to be identified with, it is a structure through which we can analyze the ways in which LGBTQ rights are often framed and deployed
- in Terrorist Assemblages, Jasbir Puar argued against the notion that the state is solely heteronormative and excludes queer individuals at all times and in all ways
    - in Rethinking Homonationalism, Puar goes on to say that homonationalism is exemplified by how queer acceptance has become "a barometer by which the right to and capacity for national sovereignty is evaluated"
        - this means that in some way or other, all LGBTQ rights campaigns uphold some ideals of social progress and act on the assumption that western nation states are capable of accepting and benefiting all people from all groups
        - Puar argues that this is impossible as these states are built upon granting privileges to certain groups, necessitating that they be taken away from others
            - e.g. the fight for and granting of marriage equality legitimize and entrench the normative ideal of marriage between two people
            - e.g. a state granting trans citizens the right to change their identity on a passport legitimizes the state's ability to control and surveil its citizens, to control movement between borders, and to deport racialized minorities
- it is difficult not to recreate homonationalist discourses whenv discussing current events such as Trump's push to ban trans servicemembers from the military
    - i.e. it can be difficult to decouple support from trans people in the military from support for the military itself
        - "trans people are just as good at killing brown people overseas as cis people."
#### Welcoming the World? Hospitality, Homonationalism, and the London 2012 Olympics
- the London 2012 Olympics were framed as a celebration of diversity and inclusion but only a specific subset of identities were included
    - inclusivity was largely limited to homonormative identities
        - e.g. middle class, "respectable" couples, affluent consumers
- The olympic games are not only a sporting event but also a massive marketing and city transformation project
    - hosting the olympics leads to the displacement of marginalized communities
        - these are generally the communities that are advertised as being helped by the olympics when the decision is made to host
- homonationalism at play
    - the LGBTQ-friendly image was tied to a narrative of British exceptionalism
        - the UK was portrayed as a global leader in civility and progress
    - this positioned the UK "against" other countries that were not as accepting of LGBTQ identities
    - the narrative was also tied to Islamaphobic undertones
        - suggested that East London's Muslim communities were intolerant and needed reform
- exclusion and regulation of queer spaces
    - transgressive sexual cultures were marginalized or directly targeted by policing
    - sex workers, particularly Eastern European migrant workers, were vilified in the media as a threat
        - despite little to no evidence of an increase in trafficking or exploitation
        - crackdowns on sex work intensified leading up to the Olympic games
- neoliberal urbanism
    - the Olympics tend to be used to accelerate gentrification
        - the games are often used to justify the displacement of marginalized communities
            - this is often framed as a positive development with wording like "urban renewal" or "clean-up" efforts
    - the fantasy image of "unity in diversity" is painted over actual exclusion and urban inequality
    - sexual diversity was commodified
        - it is celebrated when profitable, erased when disruptive
