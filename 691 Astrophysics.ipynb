- A few words about Emission
    - $I_{\lambda,0}$ --> some media $dx$ --> $I_{\lambda}$ observed
    - $I_{\lambda} = I_{\lambda,0} +dI_{\lambda}$
        - $dI_{\lambda}$ will be positive (gaining intensity with distance)
    - emission coefficient $j_{\lambda}$ (sometimes $\epsilon_{\lambda}$)

            - per wavelength 
            - per solid angle being emitted

- Equation of Radiative Transfer
    - ***potentially the most important lecture of the semester***
        - "Certainly my favorite."
    - model process
        - radiation comes into a gass cloud with intensity $I_{\lambda}$
        - gass cloud has some absorption coefficient $\alpha_{\lambda}$
        - gass cloud has some emission coefficient $j_{\lambda}$
        - gass cloud has some width $dx$
        - some radiation amount $I_{\lambda}$ comes out the other side
            - $I_{\lambda} = I_{\lambda,0} + dI_{\lambda}$
                        
                - $dI_{\lambda} = ?(emission) - ?(absorption)$
            - $dI_{\lambda} = j_{\lambda}dx - \alpha_{\lambda}I_{\lambda}dx$
            - $\frac{dI_{\lambda}}{dx} = j_{\lambda} - \alpha_{\lambda}I_{\lambda}$
                        
    - **$\frac{dI_{\lambda}}{\tau_{\lambda}} = S_{\lambda} - I_{\lambda}$ the general equation of radiative transfer**
        - $\frac{dI_{\lambda}}{d\tau_{\lambda}}$ represents the change in intensity with optical depth
        - $S_{\lambda}$ is the source function
            - how much radiation is being emitted by the medium
        - $I_{\lambda}$ is the current present intensity
            - because it is negative, it represents absorption
        - tldr; how much you gain or lose per unit optical depth depends on how much you gain minus how much you lose
        - complex and non-linear, can't just integrate
            - often solved by numerical (iterative) code methods
            - e.g. the core of a star has layers with different densities, temperatures, coefficients, etc.
                - calculate for each layer, then iterate
                - $dI = (S_{\lambda} - I_{\lambda})d\tau_{\lambda}$
                - $I_n$ for any layer $n$ is $I_{\lambda,n-1} + (S_{\lambda} - I_{\lambda,n-1})d\tau_{\lambda}$
                                
                    - i.e. each layer's output becomes the next layer's input
                - easy toy model with a few layers or a better numerical model with many layers
        - **insert image of iterative model**
            - <img src="images/iterative_model.jpeg" alt="Iterative Model" width="600"/>
        - what if we assume $S_{\lambda}$ is constant?
            - not always a valid assumption
            - not always invalid
                - can be a good approximation
            1. start with $\frac{dI}{d\tau} = S - I$
                - remember I is a function of $\tau$
            2. $\frac{dI}{d\tau} + I = S$
            
            3. $(\frac{dI}{d\tau} + I)e^{\tau} = Se^{\tau}$
            
            4. $\frac{d}{d\tau}(Ie^{\tau}) = Se^{\tau}$
                - by chain rule
            5. $d(Ie^{\tau}) = Se^{\tau}d\tau$  
                        
                - remember S is constant in this case
                - $\int_{\tau = 0, I = I_0}^{\tau, I}$
                                
            6. $Ie^{\tau}|_{0,I_0}^{\tau,I} = Se^{\tau}|_{0}^{\tau}$
            7. $Ie^{\tau} - I_0 = S(e^{\tau} - 1)$
             8. $I_{\lambda}(\tau_{\lambda}) = I_{\lambda,0}e^{-\tau_{\lambda})} + S_{\lambda}(1-e^{-\tau_{\lambda}})$
                        
                - $I_{\lambda}(\tau_{\lambda})$ is the intensity after traveling through some optical depth
                                
                - $I_{\lambda,0}$ is the initial intensity 
                    - the $e^{-\tau_{\lambda}}$ term is the fraction of the initial intensity that is absorbed
                                        
                - $S_{\lambda}$ is the source function
                    - how much radiation is being emitted by the medium
                    - the $1-e^{-\tau_{\lambda}}$ term represents some amount of self absorption of the emitted radiation
                - for very large $\tau_{\lambda}$, $I_{\lambda}(\tau_{\lambda}) \approx S_{\lambda}$
                    - because the initial intensity is absorbed
                - for very small $\tau_{\lambda}$, $I_{\lambda}(\tau_{\lambda}) \approx I_{\lambda,0}$
                                
                    - because the source function is absorbed                      
                - e.g. cases of thermal equilibrium
                    - constant S $\rightarrow$ constant Planck Function $B_{\lambda}(T)$
                    - e.g. roughly isothermal media such as planetary atmosphere
            - advantage here is that you could do it on paper and don't need a computer  
            - dummy check is case of $\tau_{\lambda} = 0$
                - $I_{\lambda}(\tau_{\lambda}) = I_0 + S_{\lambda}(1-1) = I_0$
            - other check is case of $\tau_{\lambda} >> 1$
                - $I_{\lambda}(\tau_{\lambda}) = 0 + S_{\lambda}(1-0) = S_{\lambda}$
            - for very small $\tau_{\lambda}$, $I_{\lambda}(\tau_{\lambda}) \approx I_0$

### Transition Rates & Line Profiles

- spectral lines correspond to transitions
    - exist in the plots of $I_{\lambda}, F_{\lambda}, \sigma_{\lambda}, \tau_{\lambda} vs \lambda or wavelength \nu$
    - come in different shapes
    - line will have some center $\lambda_0$ or $nu_0$
    - may be very shallow or very deep
    - may be very narrow or very broad
        - broadening effects may be due to 
            - quantum effects (uncertainty)
            - pressure
            - doppler broadening
                - rotational broadening
                    - a very quickly rotating star will have broader lines than a slower one
                - thermal effects
            - instrumental limitations
    - **Line Shape Function, $\phi(\lambda)$** 
        - a normalized function describing the general shape of a spectral line
        - defined by $\int_{-\infty}^{\infty} \phi(\lambda)d\lambda = 1$ 
            - units of $m^{-1}$ or $\mu m^{-1}$
        - can also be defined by $\int_{-\infty}^{\infty} \phi(\nu)d\nu = 1$
            - same function, just in frequency space
            - units of $Hz^{-1}$ 
        - process:
            - find the line center $\lambda_0$ for each line in the spectrum
                - "zero broadening" case
            - convolution of the line center with the line shape function
                - find the shape function based on the combined effects of all broadening processes
                - "broadening" case

- Transition Rates
    - given some lower energy state and an upper energy state
        - really you'll have lots of states but it can be modeled as two
    - any excited state will eventually de-excite/decay
    - generally this is spontaneous but there are physical rules
    - usually an average time to emit is known or found
        - **$<t_{emit}>$** 
    - Einstein Coefficients
        - $A_{UL} = \frac{1}{t_{emit}}$
            - the probability per time of a decay from the upper state to the lower state
            - thus it is also the probability per time of a photon emission
- Emission
    - take a hydrogen atom
        - has a very strong UV line, the "Lyman Alpha" line
        - $\lambda \approx 1200Ã…$
        - $<t_{emit}> \approx 2 \cdot 10^{-9}s$
        - $E_{Lyman Alpha} \approx 10.2eV$
        - $A_{UL} = \frac{1}{t_{emit}} \approx 5 \cdot 10^8 s^{-1}$
        - amount of hydrogen to see a decay in 1 ns
            - $n_{H} = \frac{1}{A_{UL}} \approx 2 \cdot 10^{-9}$ atoms
        - remember emission coefficient, $j_{\lambda}$ 
            - in $W/m^3/\mu m/sr$
            - combine:
                - $E_{\gamma} = \frac{hc}{\lambda} = h\nu$ [J[
                    - it was E
                - $A_{UL} = \frac{1}{t_{emit}}$ [s^-1]
                - $n_u =$ number density of excited atoms [m^-3]
                - $\phi(\lambda) =$ line shape function [$\mu m^-1$]
                - $\Omega = 4\pi$ solid angle [sr] 
            - $j_{\lambda} = \frac{E_{\gamma}A_{UL}n_u\phi(\lambda)}{4\pi}$
                - j is a macroscopic shorthand for the microscopic process listed above
- Absorption
    - $B_{LU}$
        - the probability per time of a decay from the lower state to the upper state
    - similar to above, the absorption coefficient, $\alpha_{\lambda}$ is made up of the same components
    - $\alpha_{\lambda} = \frac{E_{\gamma}B_{LU}n_l\phi(\lambda)}{4\pi}$
        - $B_{LU}$ is the Einstein Coefficient for absorption

- **Angular momentum L is quantized**
    - $L = \hbar n$ where $n$ is an integer
- potential energy $U_e = \frac{-k_eq^2}{r}$
    - $k_e$ is the Coulomb constant
    - $q$ is the charge of the electron
    - $r$ is the distance from the nucleus
- kinetic energy $K_e = \frac{1}{2}m_ev_e^2$
    - $m_e$ is the mass of the electron
    - $v_e$ is the velocity of the electron
- Force $F = \frac{k_eq^2}{r^2}$
- $kq^2mer^2 = m_e^2v_e^2r^2 = L^2 = (n\hbar)^2$
    - $v_e$ is the velocity of the electron
    - $r$ is the radius of the orbit
    - $L$ is the angular momentum
    - $n$ is the quantum number
- $r_n = \frac{n \hbar^2}{m_ek_eq^2}$
    - the Bohr Radius of an atom with quantum number $n$
    - Hydrogen
        - $r_H = \frac{1\cdot \hbar^2}{m_ek_eq^2}$
        - $r_H \approx 0.05nm$
- Virial theorem
    - $2K = -U$
        - we know $E_{tot} = K + U$
        - $\rightarrow E_{tot} = \frac{-1}{2}U$
- Energy of a hydrogen atom
    - $E \propto \frac{-1}{r^2}$
    - $E_H(n) = -\frac{(k_eq^2)^2m_e}{2n^2\hbar^2}$
        - n is the only non-constant term
        - **$E_H = \frac{-13.6eV}{n^2}$**
            - the negative sign indicates that the electron is bound
            - if n $\rightarrow \infty$, $E_H \rightarrow 0$
            - i.e. if n gets big, the electron becomes unbound
            - -13.6eV is the ionization energy of hydrogen
                - if you can add 13.6eV to a hydrogen atom, you can free the electron
            - all possible energy levels of hydrogen can be found by this equation by inputting the allowed n values

```python
# plot the first 5 energy levels of hydrogen
import numpy as np
import matplotlib.pyplot as plt

def energy(n):
    return -13.6/n**2

n = np.arange(1,6)
e = energy(n)

# plt.plot(n,e, 'o')
#plot a line at each energy level
for i in range(5):
    plt.plot([0,6],[e[i],e[i]], label=f'n={n[i]}')
    # label each line to the right of the lines
    plt.text(6,e[i], f'n={n[i]}', verticalalignment='center')
    # label the energy level to two decimal points to the left of the lines
    plt.text(0,e[i], f'{e[i]:.1f}', verticalalignment='center', horizontalalignment='right')
# do not draw the top or right lines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xlabel('n')
plt.ylabel('Energy (eV)')
plt.title('First 5 Energy Levels of Hydrogen')
plt.show()
```

## Molecular Transitions
- in atoms
    - electronic transitions
        - electrons moving between energy levels
- in molecules
    - also some electronic transitions
        - more complex than atoms
    - rotational
        - molecule spins around its center of mass
        - changes in angular momentum
    - vibrational
        - molecule vibrates
        - the length of the bonds are not constant
            - toy model: balls on a spring
        - changes in potential energy

### Vibration
- some $m_1$ and $m_2$ connected by a spring with spring constant $k$
    - fundamental quantities in a harmonic oscillator
        - potential energy $E = \frac{1}{2}kx^2 + \frac{1}{2}mv^2$
        - $\omega = \sqrt{\frac{k}{m}}$
            - the frequency of the vibration
        - $x = A\sin(\omega t)$
        - $F = -kx$
            - Hooke's Law
    - **in a molecule things are different**
        - $\omega_0 = \sqrt{\frac{k}{\mu}}$
            - **the fundamental frequency in rad/s**
            - $\mu$ is the reduced mass of the molecule
            - $\mu = \frac{m_1m_2}{m_1+m_2}$
        - $\nu_0 = \frac{\omega_0}{2\pi}$
            - **the fundamental frequency in Hz**
        - $E = h\nu_0(l+\frac{1}{2}) = \hbar\omega_0(l+\frac{1}{2})$
            - $l$ is the vibrational quantum number
            - $l = 0,1,2,...$
            - $\rightarrow$ even at l=0 (the ground state) there is still some vibration and thus energy
            - $\rightarrow$ the energy levels are quantized
        - e.g. the CO molecule
            - $\nu_{CO} \approx 7.5 amu$
            - $k_{CO} \approx 1800 N/m$
            - $\delta E = h\nu_0 = \frac{hc\nu_0}{\lambda}$
                - $\delta E = E_{upper} - E_{lower}$
                - $\delta E = h\nu_0$
            - $\omega_0 = \sqrt{\frac{k_{CO}}{\mu_{CO}}} \approx 2\pi \cdot 4.5 \cdot 10^{14} Hz$
            - $\nu_0 = \frac{\omega_0}{2\pi}  \approx 2/3 \cdot 10^{14} Hz$
            - $\lambda = \frac{c}{\nu_0} \approx 4.5 \mu m$
            - for $l_{0\rightarrow 2}$
                - $\delta E = 2h\nu_0$
                - $\lambda = 2\lambda$
                - $\lambda \approx 2.25 \mu m$

## Rotation
- the moment of inertia, $I$
- $K_{rot} = \frac{1}{2}I\omega^2 = \frac{1}{2}\frac{L^2}{I}$
    - $L$ is the angular momentum
    - $\omega$ is the angular velocity
- $E = \frac{1}{2}(\frac{J(J+1)}{I})\hbar^2$
    - $J$ is the rotational quantum number
    - $J = 0,1,2,...$
- rotation is easier to induce than vibration or electronic transitions
    - electronic transitions are on order $E \approx 1-2eV$
        - optical wavelengths
    - vibrational transitions are on order $E \approx 0.1-1eV$
        - Near infrared wavelengths
    - rotational transitions are on order $E \leq 1eV$
        - microwave to infrared wavelengths
        - radio wavelengths
