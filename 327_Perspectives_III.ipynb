### HIV Is Not for Me
- HIV/AIDS in WSW communities of sub-Saharan Africa are not well understood 
- research and discussions of this topic are complicated by social stigmatization of WSW individuals from within their own communities
    - WSW is often seen as not at risk for HIV/AIDS and therefore any infection is regarded as a result of promiscuity outside of "real" lesbianism
    - in reality, WSW individuals are at risk for HIV/AIDS and a lack of awareness increases the risk of infection
    - over 50% of participants had not talked to a healthcare workerd about sexual health
- increased awareness and education is needed to combat the spread of HIV/AIDS in WSW communities
    - social stigmas must be addressed to allow for open discussion and education
    - education must be tailored to the specific needs of WSW individuals
    - the internet is a valuable resource for providing information and support and some participants had used it to find information
- the African Charter on Human and People's Rights on the Rights of Women
    - ratified by most African countries
        - enforcement varies by country as does equitable access to healthcare
    - provides a framework for addressing women's rights and health issues including HIV/AIDS
    - does not specifically address sexual orientation
        - still, WSW clearly qualify as "women" and are therefore protected under the charter as women
        - WSW sex may not be considered sex within the legal definition of the charter
### From Health Crisis to Rights Advocacy
- much has been written about the relationship between HIV/AIDS and LGBTQ activism in the United States
- this paper looks into similar movements in China and Singapore
    - despite authoritarian governments, HIV/AIDS created an opportunity for LGBTQ activism
    - governments were eventually forced to address the epidemic
    - some organizations were able to use the epidemic to push for LGBTQ rights while others attempted to distance themselves from it to avoid further stigmatization
- in China, funding for HIV/AIDS research and prevention came from both international and domestic sources but was allocated by the government in either case
- in Singapore, the government controlled the funding and engagement with activists to limit their ability to challenge the political status quo
    - same-sex contact between men is illegal in Singapore
    - in the first decade, NGOs were forced to be funded domestically by local private supporters
        - Singapore is considered "developed" and therefore largely ignored by international funding sources
        - the government did not provide significant funding for HIV/AIDS research and prevention
    - in the 1990s, the government provided increased funding but used it as a means to control and limit the activities of NGOs
        - those who accepted funding were required to follow government guidelines
        - they could not advocate for other LGBTQ rights
- main strategies employed by groups
    - partner with the state, avoid LGBTQ advocacy
        - registered
        - focused on HIV/AIDS prevention
        - distanced themselves from LGBTQ activism
    - partner with the state, attempt to advocate for LGBTQ rights where possible
        - registered
        - focused on HIV/AIDS prevention
        - attempted to advocate for LGBTQ rights in non-confrontational ways
    - avoid the state, advocate for LGBTQ rights
        - unregistered
        - focused on adancing LGBTQ rights
    - avoid the state challenge the state
        - unregistered
        - campaigned against anti-LGBTQ laws
- international agencies and donors attempted to support activism in Singapore but were limited by the government's control over funding

- snow day makeup on Asexuality
    - asexuality: the orientation in which individuals do not experience sexual attraction
    - the normalization of sexual attraction
        - key concepts
            - heteronormativity
                - we assume not only straightness, but also that everyone is sexual
            - "compulsory sexuality"
                - the idea that everyone should be sexual
                - coined by Adrienne Rich
                - negative impacts
                    - pathologization
                        - lack of interest in sex has historically been pathologized as a medical condition
                        - legitimizes medical intervention, discrimination, etc.
                    - isolation and invisibility
                        - asexual individuals are often isolated and invisible in society
                    - unwanted sex and relationship conflict
                        - asexual individuals frequently feel pressured into sexual activity
                        - this can make maintaining relationships difficult
                        - in a nuanced and complex way, one may consent to sex without desiring it
                            - analogy: I don't want to help someone clean their garage, but if I love them I will because it is important to them
                    - denial of epistemic authority
                        - asexual individuals are often told that they aren't really asexual, haven't met the right person, are late bloomers, etc.
                        - denies that the asexual individual knows themselves
            - sex positivity
                - moving away from shame and stigma
                - legitimizing being a sexual being and sex for the sake of pleasure
                - can still act as a normative force analogous to homonormativity
                    - can erase asexuality if it treats sexuality as compulsory
                    - can define "normal" sexuality
                - came out of the "sex wars" of the 1980s and 1990s
                    - feminists debated whether things like porn and kink were liberating or oppressive
                - sex positivity an evolving concept
                    - originally centered around straight, white, cis, monogomous, etc. identities
                    - now it is (generally) more inclusive
                        - it is important to ensure that it is inclusive of asexuality and other identities
            - non-normative forms of sexual pleasure
                - kink, an absence of sexual pleasure, etc
        - challenging norms
            - language of difference and capacity
            - desexualizing the world
                - asexuality may not play a large role in one's life
                - asexuals may not see the world through a sexual lens
                    - i.e. (if I understood it correctly) they think in terms of sexuality so infrequently that it is not a significant part of their identity
            - new relationships
                - new types of relationships that are not based on sexual attraction
                - types of attraction
                    - sexual
                    - romantic
                    - emotional
                    - aesthetic
                    - intellectual (sapiosexual)
                        - attraction to intelligence, similar nerdy interests, neurodiversity, etc.
                    - and many more
                    - society teaches that all of these types of attraction are contained within sexual attraction
                    - this is not the case
                    - e.g. one may be (probably is?) strongly emotionally attracted to their best friend but not sexually attracted to them
                    - as a result one may be heterosexual but biromantic or many other combinations
            - orientation and/or identity
            - community building and visibility
                - forming communities and providing visibility can help to combat isolation and invisibility and challenge norms
