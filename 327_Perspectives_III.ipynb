### HIV Is Not for Me
- HIV/AIDS in WSW communities of sub-Saharan Africa are not well understood 
- research and discussions of this topic are complicated by social stigmatization of WSW individuals from within their own communities
    - WSW is often seen as not at risk for HIV/AIDS and therefore any infection is regarded as a result of promiscuity outside of "real" lesbianism

### From Health Crisis to Rights Advocacy
- much has been written about the relationship between HIV/AIDS and LGBTQ activism in the United States
- this paper looks into similar movements in China and Singapore
    - despite authoritarian governments, HIV/AIDS created an opportunity for LGBTQ activism
    - governments were eventually forced to address the epidemic
    - some organizations were able to use the epidemic to push for LGBTQ rights while others attempted to distance themselves from it to avoid further stigmatization
- in China, funding for HIV/AIDS research and prevention came from both international and domestic sources but was allocated by the government in either case
- in Singapore, the government controlled the funding and engagement with activists to limit their ability to challenge the political status quo
    - same-sex contact between men is illegal in Singapore
    - in the first decade, NGOs were forced to be funded domestically by local private supporters
        - Singapore is considered "developed" and therefore largely ignored by international funding sources
        - the government did not provide significant funding for HIV/AIDS research and prevention
    - in the 1990s, the government provided increased funding but used it as a means to control and limit the activities of NGOs
        - those who accepted funding were required to follow government guidelines
        - they could not advocate for other LGBTQ rights
- main strategies employed by groups
    - partner with the state, avoid LGBTQ advocacy
        - registered
        - focused on HIV/AIDS prevention
        - distanced themselves from LGBTQ activism
    - partner with the state, attempt to advocate for LGBTQ rights where possible
        - registered
        - focused on HIV/AIDS prevention
        - attempted to advocate for LGBTQ rights in non-confrontational ways
    - avoid the state, advocate for LGBTQ rights
        - unregistered
        - focused on adancing LGBTQ rights
    - avoid the state challenge the state
        - unregistered
        - campaigned against anti-LGBTQ laws
- international agencies and donors attempted to support activism in Singapore but were limited by the government's control over funding

#### Capitalism and Gay Identity (John D'Emilio)
- the 1970s saw a shift toward social acceptance of homosexuality and an easing of legal restrictions in the US
- the 1980s saw a resurgence of right wing politics
- queer history has often been discussed as if it began in the mid 20th century
    - because it was not well documented or at least not well understood before then, D'emilio argues that a mythic history was created in its place
    - the mythic history (probably inadvertently) suggests that the oppression of queer people is timeless and unchanging and that queer people have always existed as a distinct group
- challenging the myth
    - capitalism converted economically independent family units into dependent wage earners and consumers
        - this created a new form of family life
        - a family was still internally interdependent but not economically independent from the rest of society
            - in the 1800s, the wife still cooked bread but she bought the flour and the man worked for the money to buy the flour
            - i.e. the family was no longer self-sufficient and the woman was still dependent 
    - the new economics changed the definition of heterosexual norms
        - prior to this change, sex was meant for procreation and birth rates were high
            - multiple children were needed to work the farm or business and overcome high infant mortality rates
            - the societal definition of morally acceptable sexuality was built around procreation
                - i.e. it was not that heterosexuality was celebrated, rather it was that *any* form of sex out of wedlock was condemned
        - after the change, birth rates dropped and the purpose of sex changed
            - the economic changes allowed queer people to explore their sexualities in more material ways and form communities
            - notions of sexual identity began to form
                - i.e. homosexuality became a distinct identity rather than a behavior
        - WWII threw people into gender segregated environments 
            - it allowed queer people to find each other
            - it allowed people to realize their attractions and identities
    - still, broader capitalist society resisted and oppressed queer people
        - in some ways the enforcement of the oppression has merely shifted from the state to extralegal violence
            - (or maybe the state is about to pick up the torch again - not that the mob is dropping it)
        - capitalism acts to undermine the nuclear family
            - it removed the economic independence of the family unit
                - i.e. now a person or family has to work for someone else in order to survive
            - eventually it removed the economic interdependence within the family unit
                - i.e. now a person can work for the money to buy pre-baked bread on their own and skip the spouse, the farm, the kids, and the flour altogether
            - it has socialized child rearing and made it a public responsibility
        - despite this, it still mandates the nuclear family as the ideal
            - *a man and woman must marry and have children*
            - *children are the responsibility of the parents*
        - as capitalism knocks the foundation out from under the nuclear family, it scapegoats queer people for the instability of the system  

#### Scandalous Acts (Don Kulick & Charles H. Klein)
- Brazilian travesti prostitutes use "scandals" to extort more money from clients
    - a scandal is a loud argument effectively acting as a threat to make a public scene and reveal the purportedly heterosexual client's proclivities
    - an overwhelming majority of clients request to be penetrated by the travestis while maintaining that they are heterosexual
- the travestis are effectively using the opposite of a "pride" strategy to level the playing field in a similarly shame based society
    - a pride strategy would be to say "yes, I am homosexual and that is fine"
    - a scandal strategy would be to say "look at how homosexual you are and how wrong that is"
        - specifically it would be said loudly such that others could hear, with the intent of causing embarrassment and shame to the client into paying more
        - this redistributes power and money to the oppressed group
- travesti protests similarly do not focus on "pride"
    - lingerie and outrageous clothing and makeup are worn
    - signs carry modest slogans such as "Travestis are human beings" or "Travestis are citizens"
        - the scandal here is in that if the travesti is a human being and a citizen, then the society, like the client, is not acting humanely or like a good citizen 
    - at their most different and most shameless, travestis assert that they are most like everyone else
        - **this strategy refuses to accept differentiation and declare sameness**
- travesti, per this article at least, do not view themselves as women or transgender
    - still, they use estrogen and silicone to feminize themselves and use the pronoun "she" 

- Brazilian travisti are found in all cities of any size, appear on telivision, in magzines and parades, and sell out the copies of Playboy that they appear on
    - these travesti tend to be exceptions, not the rule
    - they are victims of brutality and discrimination
- as sex workers, travestis were especially hard hit by the AIDS epidemic
- Brazilian AIDs discourse from both travesti and other gay rights groups has centered around reterritorializing shame

                - wealthy capitalists own the means of production and the working class must sell their labor to survive

                - fewer children: having many children is no longer necessary for survival
                    - more children are more mouths to feed and not more hands to work
                - men leave the homestead: men are the ones selling their labor while women largely remain in the home 
        - the gay identity begins to form in the middle of this century
            - the breakdown of forced heterosexual coupling allows for exploration and for behaviors to become identities
                - i.e. one can now economically survive ***and*** go home to someone they are attracted to
    - 1900s: capitalist society
        - urbanization and birth rate decline continues
        - suburbanization
        - women begin to work outside the home
            - this further breaks down the interdependence of the procreative heterosexual family model
            - a woman may not need a man to survive and vice versa
            - especially during WWII
        - the nuclear family becomes the ideal
            - a heterosexual couple with children
                - shift away from multigenerational family homes, extended family, etc.
            - despite the fact that it is less economically incentivized/required than perhaps ever before
        - bachelorhood and spinsterhood
            - spinster originally meant a woman who could spin yarn so fast that she could support herself
            - an individual could now survive economically entirely without a spouse
                - dependence on parents, children, or a spouse is greatly lessened for many 
        - WWII
            - segregation of the sexes as men go to war and women join the workforce
                - factory barracks house women together
                - the military housed men together
                - spurred civil rights movements
        - the gay identity really takes shape in this period

### 28Feb25 Class
- close reading: fine grain analysis of a specific topic, section, or word in a text
- disidentification: "a descriptive of the survival strategies the minority subject practices in order to negotiate a phobic majoritarian public sphere that continuously elides or punishes the existence of subjects who do not conform to the phantasm of normative citizenship"
    - break it down:
        - the "minority subject" 
            - the group of least power who calls outside the norm
        - "phobic majoritian public sphere" 
            - the normative public society which oppresses the minority subject to enforce the normative ideal
        - the "phantasm of normative citizenship" 
            - the normative ideal
            - it is phantasmal because it isn't real, tangible, or attainable 
    - in english: disidentification describes the behaviors which marginalized people must undertake to survive more comfortably in a normative society that they do not otherwise conform to
- subject construction vs ideological practices
    - identification, counteridentification, disidentification
        - forming an identity, defining what one is, inherently requires counteridentification, defining what one isn't
        - identification generally involves evaluating a laundry list of different markers/behaviors/etc to decide whether they do or do not represent us and our desired identity
        - disidentification is then about redefining and re-contextualizing these identifying markers 
            - an active and deliberate choice 
            - incorporates both identification and disidentification
            - instead of trying to counter the toxic parts of an identity marker, it acknowledges them and attempts to transform them into something positive or otherwise more accurate
