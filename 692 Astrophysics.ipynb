### 22Jan Special Relativity
- describes motion in an inertial reference frame moving at a significant fraction of c
    - a "significant fraction" is a few percent of c
- relevant in cases such as non accelerating emission sources
    - e.g. AGN jets, gamma ray bursts
- effects include
    - apparent ftl motion
    - boosted energy of emitted photons
    - false estimates of total energy output 
    - incorrect estimate of duration of events
- postulates
    1. the equations of motion are the same in all inertial reference frames
    2. the speed of light in a vacuum is constant and universal
- consider two reference frames S and S'
    - the unprimed frame can be considered at rest
        - it is not actually at rest (no such thing) but we treat it as such
    - the primed frame is moving at a constant velocity v in the x direction
    - creates length contraction and time dilation
    - converting between the two frames
        - $x^2 + y^2 + z^2 = c^2t^2$ and $x'^2 + y'^2 + z'^2 = c^2t'^2$
        - $x' = a_{1}(x - vt)$
        - $y' = y$ 
            - unchanged because we are only considering motion in the x direction
        - $z' = z$ 
            - unchanged because we are only considering motion in the x direction
        - $t' = a_{2}t + a_{3}x$
        - the constants $a_{1}$, $a_{2}$, and $a_{3}$ encode the relativistic effects
            - for S = S', $a_{1} = a_{2} = a_{3} = 1$
    - definitions
        - $\beta = v/c$
            - $0 \leq \beta < 1$
        - $\gamma = (1 - \frac{v^2}{c^2})^{-1/2} = (1 - \beta^2)^{-1/2} =$ Lorentz factor
            - $\gamma \geq 1$
    - transformations revisited
        - $x' = \gamma(x - \beta ct)$
        - $y' = y$
        - $z' = z$
        - $ct' = \gamma(ct - \beta x)$    

- Classical limit
    - $x' \approx (x - vt)$
    - $ct' \approx ct$
- time dilation
    - $ct' = \gamma(ct - \beta x)$
    - $ct = \gamma(ct' + \beta x')$
    - consider two events at $t_{1}'$ and $t_{2}'$ in S'
        - what is the time difference between the events in S?
        - assume $x_{1}' = x_{2}' = 0$, i.e. the events are at the same location in S'
        - $t_{1} = \gamma(t_{1}' + \frac{\beta x_{1}'}{c})$
        - $t_{2} = \gamma(t_{2}' + \frac{\beta x_{2}'}{c})$
        - $t_{2} - t_{1} = \gamma(t_{2}' - t_{1}' + \frac{\beta x_{2}' - \beta x_{1}'}{c})$
        - $t_{2} - t_{1} = \gamma(t_{2}' - t_{1}' + 0)$
        - $\Delta t = \gamma \Delta t'$
    - because $\gamma \geq 1$, the clock in S' runs slower than the clock in S

### 24Jan Special Relativity Continued
- length contraction
    - $x' = \gamma(x - \beta ct)$
    - $x = \gamma(x' + \beta ct')$
    - consider two positions at $x_{1}'$ and $x_{2}'$ in S'
        - what is the distance between the positions in S at $t_{1} = t_{2} = t$?
        - $x_{2}' - x_{1}' = \gamma(x_{2} + \beta ct_{2} - x_{1} - \beta ct_{1})$
        - $x_{2}' - x_{1}' = \gamma(x_{2} - x_{1})$
        - $\Delta x = \frac{\Delta x'}{\gamma}$
    - the faster an object moves, the shorter it appears to be
    - e.g. meter stick stationary in S', S' moving at 0.5c w/r/t/ S. What is the length of the meter stick in S?
        - $\Delta x' = 1m$
        - $\gamma = (1 - \frac{v^2}{c^2})^{-1/2} = (1 - 0.5^2)^{-1/2} = (1 - 0.25)^{-1/2} = (0.75)^{-1/2} = 1.155$
        - $\Delta x = \frac{1}{1.155} = 0.866m$
- space-time invariant
    - $x^2 + y^2 + z^2 - c^2t^2 = x'^2 + y'^2 + z'^2 - c^2t'^2 = -s^2$
        - $s^2$ is the "invariant" and the negative sign is a convention
    - the invariant is the same in all inertial reference frames
    - e.g.
        - for simplicity, set $y' = z' = 0$
        - $x = 1$, $y = z = t = 0$
        - we know that $x' = \gamma(x - \beta ct)$
            - $x = 0$, $t = 0$
        - $x' = \gamma = (1 - \frac{v^2}{c^2})^{-1/2}$
        - t' = $\gamma(t - \frac{\beta x}{c}) = - \frac{\gamma \beta x}{c}$
        - $x'^2 + y'^2 + z'^2 - c^2t'^2 = x'^2 - c^2t'^2 = ?$
        - $\gamma^2 - \frac(gamma^2 \beta^2 x^2}{c^2} = \gamma^2(1 - \beta^2) = \frac{1}{1 - \beta^2} - \frac{1}{1 - \beta^2} = 1$
        - hence $x^2 + y^2 + z^2 - c^2t^2 = x'^2 + y'^2 + z'^2 - c^2t'^2 = -s^2 = 1$
- space-time invariant intervals
    - consider 2 events with different space-time coordinates
    - $\Delta x = x_{2} - x_{1}$, $\Delta x' = x_{2}' - x_{1}'$
    - $\Delta t = t_{2} - t_{1}$, $\Delta t' = t_{2}' - t_{1}'$
    - $(\Delta x)^2 - c^2(\Delta t)^2 = - (\Delta s)^2$
        - can be used to tell if two events could be causally connected
        - if $\Delta s^2 = 0$, the events are light-like separated
            - the interval is the same as that of a light ray traveling at c
        - if $\Delta s^2 \lt 0$, the events are space-like separated
            - $\Delta x \gt c \Delta t$
            - light would not have time to connect the events 
        - if $\Delta s^2 \gt 0$, the events are time-like separated
            - $\Delta x \lt c \Delta t$
            - light would have time to connect the events
- proper time interval
    - for a timelike interval, $\Delta \frac{s}{c} = \Delta t$
    - $\Delta t$ is the proper time interval
        - $(\Delta t)^2 = (\Delta t')^2 - (\frac{\Delta x'}{c})^2$
    - consider two events with $\Delta x = 0$, i.e. the same place in S'
        - $(\Delta t) = (\Delta t') = \frac{\Delta t'}{\gamma}$
        - this means that the proper time interval is the time interval in the special frame in which the events are in the same position
        - in any other frame, the time interval will be longer
- four-vector
    - a vector to describe a spacetime position
    - $[x, y, z, t] = [\mathbf{X}, t]$ where $[\mathbf{X}, t]$ is the position vector
    - $[x, y, z, ct] \cdot [x, y, z, ct] = x^2 + y^2 + z^2 - c^2t^2$
        - not a dot product
- determining relativistic momentum
    - $\frac{m}{\Delta \tau} = \frac{\gamma m}{\Delta t}$
        - or differentially, $\frac{\gamma m}{dt}$
        - can multiply the differential four-vector $[dx, cdt]$ by $\frac{\gamma m}{dt}$ to get the relativistic momentum four-vector
    - $[dx, cdt] \times \frac{\gamma m}{dt} = [\gamma m \frac{dx}{dt}, \frac{\gamma m c^2}{c}] = [\gamma m v, \gamma m c] = [\mathbf{p}, \frac{U}{c}]$
        - the first component is the relativistic momentum
            - $\mathbf{p} = \gamma m \mathbf{v}$
        - the second component is the energy
    - the energy term of the energy-momentum four-vector has a $\gamma m c^2$ term
        - $\gamma m c^2 = (1 - \frac{\nu^2}{c^2})^{-1/2} m c^2$
            - do a taylor expansion to get the classical limit
        - $\gamma m c^2 = m c^2 + \frac{1}{2} m \nu^2 + ...$
        - for $\nu \lt \lt c$, $\gamma m c^2 \approx m c^2 + \frac{1}{2} m \nu^2$
        - $U = \gamma m c^2 = E_{kinetic} + E_{rest}$
        - $E_{rest} = m c^2$
        - $E_{kinetic} = \gamma m c^2 - E_{rest}$
        - can define $\gamma = \frac{E_rest + E_{kinetic}}{mc^2}$
        - as $\nu \rightarrow c$, $\gamma \rightarrow \infty$
            - this is why you can't accelerate a massive object to c


### 27Jan
- momentum-energy relation and photon momentum
    - $U^2 - (pc)^2 = (mc^2)^2$
        - the invariant expression for momentum and energy
        - the fundamental relationship between total energy, momentum, and rest mass
        - this is the same in all reference frames
    - photons don't have mass and can never be at rest
        - $U_{photon} = pc = h\nu \implies p = \frac{h \nu}{c}$
            - photons have momentum that increases with energy
                - can be transferred to other particles
- relativistic doppler shift for source moving along the line of sight
    - e.g.
        - two frames
            - S: source at rest
            - S': source moving at velocity v relative to S
        - an observer at O in S seeing a light flashing at O' with interval $\Delta t'$ in S'
        - two factors
            - the distance traveled by the light in the time interval $\Delta t'$ in S'
                - $c \Delta t'$
            - the time dilation of subsequent flashes
                - $\Delta t = \gamma \Delta t'$
        - the difference between flash 1 and flash 2 is $\Delta x = (c - v) \Delta t$
        - to get to the observer, flash 2 has to travel $\Delta x$ further, taking $\Delta \frac{x}{c}$ longer than flash 1
        - the two flashes are only separated by time in S'
            - $(t_{1}', x_{1}' = (t_{1}', 0)$, $(t_{2}', x_{2}' = (t_{2}', 0)$
                - Lorentz transform to S
            - $(t_{1}, x_{1} = (\gamma(t_{1}' - \gamma \beta c t_{1}')$ 
                - our book bucks convention by changing the sign convention
                    - defines receding motion as negative, $\beta \lt 0$
                - hence it takes $\frac{\gamma \beta c t_{1}'}{c} = \gamma \beta t_{1}'$ longer for the light to reach the observer
            - $t_{1, obs} = \gamma(1 - \beta)t_{1}'$
            - $t_{2, obs} = \gamma(1 - \beta)t_{2}'$
            - the time difference including time dilation is
                - $\Delta t = t_{2, obs} - t_{1, obs} = \gamma(1 - \beta) \Delta t' = (\frac{1-\beta}{1+\beta})^{1/2} \Delta t'$
                - the observer sees the flashes separated by a longer time interval

        - math
        - $\frac{\nu_{obs}}{\nu_{emitted}} = (\frac{1+\beta}{1-\beta})^{1/2}$
        - $\frac{\lambda_{obs}}{\lambda_{emitted}} = (\frac{1-\beta}{1+\beta})^{1/2}$

- Doppler shift for a source moving along the line of sight
    - $\frac{\nu_{obs}}{\nu_{emitted}} = \frac{1}{\gamma(1 - \beta cos \theta)} = \frac{(1-\beta)^{1/2}}{\gamma(1 - \beta cos \theta)}$
        - $\theta$ is the angle between the line of sight and the direction of motion
        - $\beta = v/c$
        - $\gamma = (1 - \beta^2)^{-1/2}$
        - the $\gamma$ accounts for the time dilation
            - i.e. the observer sees the flashes separated by a longer time interval than the source emits them
        - the $(1 - \beta cos \theta)$ accounts for the the compression of the waves as the source moves towards the observer
            - i.e. the more classical model of Doppler shifting
    - when $\theta = 90^{\circ}$, there is still a doppler shift due to time dilation
        - $\frac{\nu_{obs}}{\nu_{emitted}} = \frac{1}{\gamma} = (1 - \beta^2)^{1/2}$
        - e.g. for an observer looking at a rotating disk, you still see a doppler shift at the nearest and furthest points
        - e.g. things moving slightly toward you may not have a doppler shift as time dilation counteracts the compression of the waves
    - when $\theta = 0^{\circ}$, the doppler shift is the largest
        - $\frac{\nu_{obs}}{\nu_{emitted}} = \frac{1+\beta}{(1-\beta)^{1/2}}$ 


- AGN jets
    - relativistic effects
        - beaming (aberation)
            - changes the angle of emission
            - $cos \theta = \frac{cos \theta' + \beta}{1 + \beta cos \theta'}$
                - $\theta$ is the angle of emission in the observer's frame
                - $\theta'$ is the angle of emission in the source's frame
                - for $\theta' = 0$, $cos \theta = \frac{1 + \beta}{1} = 1 + \beta$
                - for $\nu \rightarrow c$, $cos \theta \rightarrow 1$
                    - close to c, all light will be seen emitting in the direction of motion
                    - an approaching source will appear brighter as the rays are beamed towards the observer
                    - a receding source will appear dimmer as the rays are beamed away from the observer
                - for $\theta' = 90^{\circ}$, $cos \theta = \frac{\beta}{1} = \beta$
                    - i.e. the rays are emitted perpendicular to the direction of motion (assumes motion is towards the observer)
            - beaming angle $\theta_{b}$ is a half angle cone in the observer frame containing half the rays from the emitted frame
                - $2 \theta_{b}$ is the width of the cone (because it is a half angle)
                - for $\beta = 0.99$, $\theta_{b} = cos^{-1}(\frac{1}{\gamma}) = 8.11^{\circ}$
                    - that's a really small cone to contain half the rays
                    - e.g. at the CMS, when they do collisions, even the particles that spray out at 90 degrees are still within a 1 degree cone forward in our frame
                - $\theta_{b} \approx \frac{1}{\gamma}$ in radians
        - doppler boosting
            - the energy of the photons is boosted by the motion of the source
                - doppler factor $\delta = \frac{\nu_{obs}}{\nu_{emitted}}$
                - for $\theta = 0 \implies \delta = \frac{1+\beta}{(1-\beta)^{1/2}}$
                - $\delta = \gamma(1 + \beta)$
                - tells you how much the energy of the photons is boosted
                - e.g. (figure 7.10) an isotropic emitter at $\gamma = 10$
                    - the small cone is narrowed to $\theta_{b} = 6^{\circ}$
                        - this contains all the light from the forward direction
                        - within this cone, there is extreme boosting and blue shifting
                    - deriving the angle of the larger cone
                        - $\delta = 1 = \frac{(1-\beta^2)^{1/2}}{1-\beta cos \theta}$
                            - we must find $\beta$ from $\gamma = 10$ 
                        - $\gamma = (1 - \beta^2)^{-1/2} = 10$
                        - $\beta^{-2} = (1 - \beta^2)$
                        - $\beta = \sqrt{1 - \frac{1}{100}} = 0.995$
                            - plug in to $\delta = 1 = \frac{(1-\beta)^{1/2}}{1-\beta cos \theta}$
                        - $1-\beta cos \theta = (1-\beta^2)^{1/2}$
                        - $cos \theta = \frac{1-(1-\beta^2)^{1/2}}{\beta}$
                        - $cos \theta = 0.905 \approx 25^{\circ}$
                    


### 31Jan
- beaming vs doppler boosting
    - beaming derives from aberration
        - it's direction is changed from the source reference frame
    - boosting is a change in energy and frequency
        - it's energy is changed from the source reference frame
- Doppler boosting
    - intensity is boosted by a factor of $\delta^3$
        - due to several factors
            - more energy per solid angle (factor of $\delta^2$)
            - doppler shift of E (factor of $\delta$)
            - time dilation (factor of $\delta$)
    - however, bandwidth is increased in the observer frame, reducing the flux per frequency
        - the observer sees a wider range of frequencies
    - $I(\nu, \theta, \phi, t) = \delta^3 I'(\nu', \theta', \phi', t')$
    - calculating the solid angle in the observer frame
        - $d\Omega' = sin \theta' d\theta' d\phi'$
            - $\implies math$ 
        - $d\Omega' = (\frac{(1-\beta^2)}{(1-\beta cos \theta)^2})sin \theta d\theta d\phi = \delta^2 d\Omega$
        - for convenience, we define $\phi = \phi'$ so that only $\theta$ changes
            - i.e. phi is perpendicular to the direction of motion
    - calculating the doppler component
        - $\delta t_{obs} = (1-\beta cos \theta) \gamma \delta t'$
        - time is compressed in the observer frame
    - relativistic K-correction
        - how spectrum shifts because of doppler effect
        - log space 
            - spectra are stretched in linear space but shifted in log space
            - in classical doppler shift, each frequency is shifted by a constant amount
            - in relativistic doppler shift, the shift is proportional to the frequency
            - "by the magic of logarithms", in log space, it the spectrum is shifted not stretched
            - the spectrum will have the same shape, it is just shifted in frequency and intensity
