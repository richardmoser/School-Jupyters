### APIs
- Application Programming Interface
    - offers a service to other pieces of software
    - computer to computer, not human to computer
- a program calls a part of the API
    - calls are refered to as subroutines, methods, requests, endpoints, etc
- generally not language sensitive
- categories
    - libraries (remote)
        - can allow programs to use libraries in another language
    - frameworks
    - Operating System
    - remote
        - allow remote 4
    - Web API
        - handle interactions between enterprise systems and apps that use their assets

### Protocols
- REST
- SOAP
- RPC

### XML
- Extensible Markup Language
- tag based 
    - looks like HTML
- stores and transfers data

### REST
- probably the best choice for a modern API
    - fully leverages World Wide Web standards
    - simpler than SOAP
    - not highly coupled like RPC
- not a particular format as much as a set of design constraints
- generally use HTTP methods (GET, POST, PUT, DELETE, etc)
- generally use JSON or XML encoded data transfers
- client-server architecture
    - keep the browser from needing to know anything about how the server is designed or what languages are used
- statelesness
    - no session server is retained by the server
    - data is sent to the server by the client such that any individual packet of information makes sense in isolation
        - i.e. without the context of previous packets in the session
    - high volume operations benefit from this
        - browser can access any server and get the same result 
        - vice needing the one that "knows" you
    - cacheability
        - intermediary servers between client and destination
    
    - Advantages of caching
        - faster response
            - cahce can store a server's response to a request
                - reuse the canned response for future requests
                - don't need to interact with the web server for such requests
            - e.g. browser stores it's own cache 
        - reduces server load
            - server doesn't have to process 100% of the requests
- Layered System
    - client usually doesn't know if connected to end server or intermediary
- Uniform Interface
    - decouples architecture 
    1. resource identification in requests
        - must use URLs to identify files
    2. resource manipulation through representations
    3. self descriptive messages
        - messages include enough information to be processed
    4. hypermedia as the engine of application state (HATEOAS)
        - client should be able to use server provided links (hyperlinks) to discover the resources required
- Code on Demand

### SOAP
- old (90s kid)
- XML based
- 3 parts
    1. envelope which defines message structure and how to process it
    2. encoding rules for application defined datatypes
    3. convention for representing procedure calls and responses
    
### RPC
- older yet
- for client to execute code on the server
- highly coupled with the language
- XML-RPC uses XML
- JSON-XML uses JSON
