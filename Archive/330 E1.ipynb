````c++
#include <iostream>
#include <list>

void printLots(const std::list<int>& L, const std::list<int>& P) {
    auto lit = L.begin();  // L.begin() returns a const_iterator
    auto pit = P.begin(); // P.begin() returns a const_iterator
    int pos = 0; // position of the element in L

    while (lit != L.end() && pit != P.end()) { // while both lists have elements
//     while (lit != L.end()) { // while both lists have elements
        if (pos == *pit) { // if the current position is in P
            std::cout << *lit << " "; // print the element
            ++pit; // move to the next element in P
        }
        ++lit; // move to the next element in L
        ++pos; // increment the position
    }

    std::cout << std::endl; // print a new line
}


// int main()
// {
    std::list<int> L = {2,4,6,8,10,12,14,16,18 };
//     std::list<int> L = { 9,8,7 };
    std::list<int> P = {0, 2, 4, 8,};

    printLots(L, P);

//     return 0;
// }
````

In C++ data structures, there are several terms that refer to the start and end of a data structure, but their meaning can vary depending on the specific data structure. Here are some common examples:

- In a linked list, "head" refers to the first node of the list, while "tail" refers to the last node of the list. The "head" pointer typically points to the first node, while the "tail" pointer points to the last node.
- In a queue (e.g. std::queue), "front" refers to the element at the beginning of the queue, while "back" refers to the element at the end of the queue. The "front" element is the next element that will be removed from the queue, while the "back" element is the last element that was added to the queue.
- In a deque (e.g. std::deque), "front" and "back" have the same meaning as in a queue. However, a deque also supports adding and removing elements from both ends, so "front" and "back" can refer to either end of the deque.
- In a vector (e.g. std::vector), "begin" refers to the iterator pointing to the first element in the vector, while "end" refers to the iterator pointing one past the last element in the vector. Note that "end" is not a valid iterator and should not be dereferenced.
