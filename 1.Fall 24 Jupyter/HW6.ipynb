1. Stellar Structure – of the Earth! I. [17 pts]. Planets like the Earth are governed by some of the same structure equations that we introduced for stellar interiors. Limit your answers below to two significant figures; the crude model we’ll use below often isn’t even that accurate.
- (a) Let’s make the simplest possible model – assume that the Earth has a constant density, $\rho_{0}$. Given the known mass and radius of the Earth, calculate $\rho_{0}$. [2 pts]

Answer: 
- The mass of the Earth is $M_{\oplus} = 5.972 \times 10^{24} kg$ and the radius of the Earth is $R_{\oplus} = 6.371 \times 10^{6} m$. The volume of the Earth is $V = \frac{4}{3}\pi R_{\oplus}^{3}$. The density of the Earth is $\rho_{0} = \frac{M_{\oplus}}{V}$.
- $\rho_{0} = \frac{5.972 \times 10^{24} kg}{\frac{4}{3}\pi (6.371 \times 10^{6})^{3} m^{3}} = 5.52 \times 10^{3} kg/m^{3}$.

- (b) Assuming a constant density, calculate and plot the Earth’s enclosed mass, $M_{enc}(r)$, from $0 < r < R_{\oplus}$. What do you calculate for $M_{enc}(r = R_{\oplus})$, i.e. for the total mass of the Earth ($M_{\oplus}$)? Explain how well your answer compares to the true, measured mass of the Earth. [5 pts]

Answer:
- $M_{enc}(<r) = \int_{0}^{r}dm = \int_{0}^{r}4\pi r^{2}\rho dr$
- $M_{enc}(r) = \frac{4}{3}\pi r^{3}\rho_{0}$ for constant density
- $M_{enc}(r = R_{\oplus}) = \frac{4}{3}\pi (6.371 \times 10^{6})^{3} 5.52 \times 10^{3} = 5.972 \times 10^{24} kg$

- (c) Assuming a constant density, calculate and plot the Earth’s internal gravity field, g(r), for 0 < r < $R_{\oplus}$. Explain how well your answer for g(r = $R_{\oplus}$) compares to the true, measured surface gravity at the surface of the Earth. [5 pts]

- $g_{internal}(r) = \frac{G M_{enc}(r)}{r^{2}}$ internal gravity

```python
import numpy as np
import matplotlib.pyplot as plt

G = 6.67430e-11
R_earth = 6.371e6
M_earth = 5.972e24

def M_enc(r, rho):
    return 4/3*np.pi*r**3*rho

def g_internal(r, R_earth, rho):
    if r == 0:
        return 0
    elif r < R_earth:
        numerator = G*M_enc(r, rho)
        denominator = r**2
        frac = numerator/denominator
        return frac
    else:
        return G*M_earth/(r**2)

max_lim = 2.5 * R_earth
r = np.linspace(0, max_lim, 1000)
rho = 5.52e3
g = [g_internal(i, R_earth, rho) for i in r]

plt.axvline(x=R_earth, color='r', linestyle='--')
plt.plot(r, g)
plt.text(R_earth*1.05, 0, 'R_earth', color='r')

plt.xlim(0, max_lim)
plt.xlabel('radius (m)')
plt.ylabel('gravity (ms$^{-2}$)')
plt.title('Internal Gravity Field of Earth')
# plt.show()
plt.savefig('PS5_1c.png')

# print(f"max gravity value: {max(g)}")

#save as PS5_1c.png
pl
```

- (d) Assuming a constant density, calculate and plot the Earth’s internal pressure, P(r), for 0 < r < $R_{\oplus}$. Calculate the pressure at the very center of the Earth, $P_{c} = P(r = 0)$. [5 pts]

```python
import numpy as np
import matplotlib.pyplot as plt

G = 6.67430e-11
R_earth = 6.371e6

def M_enc(r, rho):
    return 4/3*np.pi*r**3*rho

def P_internal(r, R_earth, rho):
    if r == 0:
        return 0
    if r > R_earth:
        return 0
    else:
        return (2/3) * np.pi * G * rho**2 * (R_earth**2 - r**2)
    return (2/3) * np.pi * G * rho**2 * (R_earth**2 - r**2)

max_lim = 2.5 * R_earth
r = np.linspace(0, max_lim, 1000)
rho = 5.52e3

P = [P_internal(i, R_earth, rho) for i in r]

plt.axvline(x=R_earth, color='r', linestyle='--')
plt.plot(r, P)
plt.text(R_earth*1.05, 5000000000, 'R_earth', color='r')

plt.xlim(0, max_lim)
plt.xlabel('radius (m)')
plt.ylabel('pressure (Pa)')
plt.title('Internal Pressure of Earth')
# plt.show()

#print maximum pressure value
P_max = max(P)
print(f"max pressure value: {P_max}")


plt.savefig('PS5/PS5_1d.png')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the density profile from the CSV file
df = pd.read_csv('preliminary_reference_earth_model.csv')
# Headings are radius_meters,density_kg_per_m3
r = df['radius_meters'].values
rho = df['density_kg_per_m3'].values

# Enclosed mass calculation
def calculate_enclosed_mass(r, rho):
    # Initialize array for enclosed mass
    M_enc = np.zeros(len(r))
    # Iterate over radius points to calculate cumulative mass
    for i in range(1, len(r)):
        dr = r[i] - r[i - 1]  # Shell thickness
        M_enc[i] = M_enc[i - 1] + 4 * np.pi * r[i]**2 * rho[i] * dr
    return M_enc

# Calculate enclosed mass
M = calculate_enclosed_mass(r, rho)

# Plot enclosed mass
plt.figure(figsize=(8, 6))
plt.plot(r, M, label='Enclosed Mass')
plt.xlabel('Radius (m)')
plt.ylabel('Mass (kg)')
plt.title('Enclosed Mass of Earth')
plt.legend()
plt.grid()

M_max = M[-1]
print(f"Total mass of Earth: {M_max:.2e} kg")

# Save the plot
plt.savefig('PS5/PS5_2b.png')
plt.show()
```

Calculate and plot the Earth’s internal gravity field, $g(r)$, for $0 < r < R_{\oplus}$. The plot you get may surprise you – it did me! Describe and try to explain the general, overall features shown in your plot. Explain how well your answer for $g(r = R_{\oplus})$ compares to the true, measured surface gravity at the surface of the Earth.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the density profile from the CSV file
df = pd.read_csv('preliminary_reference_earth_model.csv')
# Headings are radius_meters,density_kg_per_m3
r = df['radius_meters'].values
rho = df['density_kg_per_m3'].values

# Constants
G = 6.67430e-11  # Gravitational constant
R_earth = 6.371e6  # Earth's radius

# Enclosed mass calculation
def calculate_enclosed_mass(r, rho):
    # Initialize array for enclosed mass
    M_enc = np.zeros(len(r))
    # Iterate over radius points to calculate cumulative mass
    for i in range(1, len(r)):
        dr = r[i] - r[i - 1]  # Shell thickness
        M_enc[i] = M_enc[i - 1] + 4 * np.pi * r[i]**2 * rho[i] * dr
    return M_enc

# Calculate enclosed mass
M = calculate_enclosed_mass(r, rho)

# Internal gravity calculation
def calculate_internal_gravity(r, M):
    # Initialize array for gravity
    g = np.zeros(len(r))
    # Iterate over radius points to calculate gravity
    for i in range(1, len(r)):
        g[i] = G * M[i] / r[i]**2
    return g

# Calculate internal gravity
g = calculate_internal_gravity(r, M)

# Plot internal gravity
plt.figure(figsize=(8, 6))
plt.plot(r, g, label='Internal Gravity')
plt.axvline(x=R_earth, color='r', linestyle='--', label='Earth\'s Surface')

plt.xlabel('Radius (m)')
plt.ylabel('Gravity (m/s^2)')
plt.title('Internal Gravity Field of Earth')
plt.legend()
plt.grid()

g_max = g[-1]
print(f"Surface gravity of Earth: {g_max:.2f} m/s^2")

# Save the plot
plt.savefig('PS5/PS5_2c.png')
```

Calculate and plot the Earth’s internal pressure, $P(r)$, for $0 < r < R_{\oplus}$. Give the central pressure, $P_{c}$, for this model and describe how it compares to the central pressure you calculated in the previous problem.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the density profile from the CSV file
df = pd.read_csv('preliminary_reference_earth_model.csv')
# Headings are radius_meters,density_kg_per_m3
r = df['radius_meters'].values
rho = df['density_kg_per_m3'].values

# Constants
G = 6.67430e-11  # Gravitational constant

# Enclosed mass calculation
def calculate_enclosed_mass(r, rho):
    # Initialize array for enclosed mass
    M_enc = np.zeros(len(r))
    # Iterate over radius points to calculate cumulative mass
    for i in range(1, len(r)):
        dr = r[i] - r[i - 1]  # Shell thickness
        M_enc[i] = M_enc[i - 1] + 4 * np.pi * r[i]**2 * rho[i] * dr
    return M_enc

# Calculate enclosed mass
M = calculate_enclosed_mass(r, rho)

# Internal pressure calculation using hydrostatic equilibrium
def calculate_internal_pressure(r, M, rho):
    # Initialize array for pressure
    P = np.zeros(len(r))
    # Iterate over radius points to calculate pressure
    for i in range(len(r) - 2, -1, -1):  # Go from the surface inward
        dr = r[i + 1] - r[i]  # Shell thickness
        P[i] = P[i + 1] + rho[i] * G * M[i] / r[i]**2 * dr
    return P

# Calculate internal pressure
P = calculate_internal_pressure(r, M, rho)

# Plot internal pressure
plt.figure(figsize=(8, 6))
plt.plot(r, P, label='Internal Pressure')
plt.xlabel('Radius (m)')
plt.ylabel('Pressure (Pa)')
plt.title('Internal Pressure of Earth')
plt.legend()
plt.grid()

# Central pressure
P_c = P[1]
print(f"Central pressure of Earth: {P_c:.2e} Pa")

# Save the plot
plt.savefig('PS5/PS5_2d.png')
plt.show()
```
