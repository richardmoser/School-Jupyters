### Netowrk Basics
- NIC
	- Network Interface Controller/network card
- Addresses
	- MAC address
		- Media Access Control address
		- Every physical port has one 
		- First half indicates the manufacturer
		- Second half should be a unique ID
			§ Hardware assigned default but can be software assigned
			§ Some devices like phones randomize the MAC every time they connect for privacy purposes
				- Large enough pool that collision is unlikely
				- We will generally assume that this isn't happening in the course
	- IP address
		- Internet Protocol Address
		- Assigned to each device in a network
		- Used for host and network ID
		- Location addressing
		- Use ifconfig to find the "inet" address
		- Versions
			§ IPV4 (32 bits for an address)
				- 2^32 combinations
				- ~4 billion
				- Need more addresses for all these devices
					® "Why would you put a pressure cooker on the internet? I guess to make life more interesting…"
				- 192.168… are private ranges
					® Ignored by big internet routing
			§ IPV6 (128 bits for an address)
				- 2^128
				- 340,282,366,920,938,463,463,374,607,431,768,211,4# 29Aug56 addresses
				- Trailing zeroes in an address can be omitted when not used
					® Often the last few digits are omitted due to large number of possible combinations
	- Differences
		- Slight oversimplification
			§ IP supports software
			§ MAC supports hardware implementation of the network stack
		- MAC generally remains fixed across networks
		- IP changes as device is moved from network to network
		- Sending a packet into the world
			§ Local network uses MAC between devices
			§ Big internet uses IP between cities/countries/whatever
			§ Goes to the destination local network which uses MAC again
- Ports
	- Physical or virtual connection points
		- Physical
			§ Allow cable connections
		- Virtual
			§ Allow virtual connections
	- Numbering
		- Well known (0-1023)
			§ 20 & 21: FTP
			§ 22: SSH
			§ 23: Telnet
			§ 25: SMTP
			§ 80: HTTP
			§ 443: HTTPS
		- Registered (1024-49151)
		- Dynamic or private (49152-65535)
		- To hide/secure a web server, use a nonstandard port so only you will know it's location

            - Null
                - send a packet with no flags set
                    - sends a TCP segment to a port with no flags set
                        - violates TCP RFC
                    - if RST is received, the port is closed
                    - if no response, the port may be open
            - TCP ACK
                - send ACK to port
                    - if RST is received, the packet made it through the filter
                    - or the port is unfiltered
                - if no response, the port is filtered

- Exploiting Vulnerabilities & Metasploit Overview
    - exploits
        - software, data chunks, sequences that take advantage of a vulnerability
        - the goal is to cause unintended behavior
            - e.g. privilege escalation, DoS, etc
        - local vs remote
            - remote occurs over a network
                - does not require any prior access
                - e.g.
                    - old Apache HTTP Server allowed remote user to cause DoS by sending a specially crafted request
            - local
                - requires prior access
                - usually involves increasing privileges beyond what is intended
                - e.g. in old Ubuntu versions, you could overflow the buffer of the "login" portion of the ftp terminal command
                    - fpt client runs at the same permissions as the user so you aren't really gaining anything
            - it is not about physical location
                - e.g. gaining super user access over SSH is considered a local exploit due to the prior access
        - exploit databases
            - websites that list known vulnerabilities
            - e.g. exploit-db.com
            - can be used to find vulnerabilities in a system
    - Metasploit
        - framework for developing and executing exploits
        - modular design
            - exploits
                - code that takes advantage of a vulnerability
            - payloads
                - ccode to run remotely once the exploit gets you in
            - auxiliary
                - tools to gather information
            - encoders
                - tools to convert from one format to another
            - and more...
        - available in the open-source Framework and Pro versions
            - Pro is **really** expensive
            - Framework is free
        - cheat sheet: https://www.ittc.ku.edu/~alexbardas/eecs465/materials/various/metasploit-cheat-sheet.pd

- DoS/DDoS
    - DoS: an explicit attempt by attackers to prevent legitimate users from using a service
    - DDoS: DoS but specifically executed by multiple users/bots in a coordinated manner
    - major types
        - vulnerability 
        - reflector
        - flooding
        - slowloris
    - e.g. a "flood"
        - put a lot of traffic on a network to prevent legitimate traffic from getting through
    - e.g. Smurf attack
        - send an ICMP echo request to a broadcast address
            - basically a "you up?" request
        - send those to a bunch of addresses with a spoofed return address
            - the return address is the target
        - drown the target in a flood of tiny replies
        - machines are not configured to ignore these requests
    - e.g. Ping of Death
        - send a packet that is too large for IP protocol
        - some machines would freeze, crash, or reboot
        - fixed by software patches
    - e.g. Nuke attacks
        - similar to the Ping of Death
        - send some other kind of malformed packet to crash the target
    - e.g. Amplification attacks
        - a type of smurf attack
        - allows a single user to use existing infrastructure to amplify/distribute their attack
        - DNS amplification
            - target open recursive DNS servers
            - spoof the source IP address (return address) to the target
            - DNS queries for domains with long responses can be obtained with a small request
            - the target is flooded with the large responses
            - several different ways to mitigate based on scenario
                - e.g. rate limiting, filtering, etc
        - memcached amplification
            - target public memcached servers
                - used to cache data and reduce database load  
            - spoof the source IP address to the target
            - receive an amplification factor of 51,200
                - 1 byte request -> 51,200 byte response
            - solution was to disable UDP by default on memcached servers
            - can also use a firewall to block UDP traffic
            - arguably the result of widespread misconfiguration
    - e.g. SYN flood
        - send a bunch of SYN packets
        - the target responds with a SYN/ACK
        - the attacker never responds with an ACK
        - the target waits for the ACK and eventually times out
        - saturates the number of available connections on the target
        - victim system cannot respond to legitimate traffic until attack stops
        - solution is SYN cookies
            - essentially expand the size of the connection queue
    - e.g. UDP flood
        - send a truly large number of UDP packets
        - target is overloaded and cannot respond to legitimate traffic
        - much harder to mitigate at the individual level
            - target rate limiting can be difficult
            - like trying to stop a flood far downriver of lots of streams
        - more effective at the collective level
            - egress filtering by ISP
                - block outgoing traffic from addresses that are not on their network
            - Content Delivery Networks (CDNs)
                - replicate content across multiple servers
                - your Netflix.com video comes from the closest server, not the main one
            - like damning streams to control flooding far down river
    - e.g. LOIC (Low Orbit Ion Cannon)
        - open source network stress testing application
        - can also be used to perform 3 types of attacks
            - TCP, UDP, and HTTP
        - each open several connections to a target and continuously send a defined payload string
        - source IP is not spoofed
            - egress filtering would have no effect since the address is legitimate

 ### DoS/DDoS Continued
- Slowloris attacks
    - goal is to obtain and keep as many open connections as possible
        - cannot be detected by a firewall because the connections are legitimate
- other DoS tools
    - Linux
        - Trin00
        - Tribe Flood Network
        - Stacheldraht
            - German for "barbed wire"
            - advanced version of TFN
            - uses encryption between master and slave
        - the above work as large master-slave botnets
            - attacker installs a master on a machine or machines
            - master sends commands to the slaves (daemons)
            - slaves send the attack to the target
    - botnets
        - collection of compromised machines controlled by an attacker
            - centralized
                - all bots connect to a central server
                - easier to take down
                - easier to manage
            - decentralized
                - there is no single master machine
                - each bot can act as a master and a slave
                    - connected through p2p
                - very hard to take down
        - e.g. Mirai
            - 200k-300k bots
            - built by 3 college students
            - goal was to do testing on a small scale network
            - targets IoT devices
            - responsible for KrebsOnSecurity and Dyn attacks
        - e.g. Nugache
            - installed through a trojan
            - computed a list of peers from a seed
            - used P2P for almost all of C&C, updates, membership, etc
            - active from 2006-2008
