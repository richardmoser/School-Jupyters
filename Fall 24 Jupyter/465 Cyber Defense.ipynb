### Netowrk Basics
- NIC
	- Network Interface Controller/network card
- Addresses
	- MAC address
		- Media Access Control address
		- Every physical port has one 
		- First half indicates the manufacturer
		- Second half should be a unique ID
			§ Hardware assigned default but can be software assigned
			§ Some devices like phones randomize the MAC every time they connect for privacy purposes
				- Large enough pool that collision is unlikely
				- We will generally assume that this isn't happening in the course
	- IP address
		- Internet Protocol Address
		- Assigned to each device in a network
		- Used for host and network ID
		- Location addressing
		- Use ifconfig to find the "inet" address
		- Versions
			§ IPV4 (32 bits for an address)
				- 2^32 combinations
				- ~4 billion
				- Need more addresses for all these devices
					® "Why would you put a pressure cooker on the internet? I guess to make life more interesting…"
				- 192.168… are private ranges
					® Ignored by big internet routing
			§ IPV6 (128 bits for an address)
				- 2^128
				- 340,282,366,920,938,463,463,374,607,431,768,211,4# 29Aug56 addresses
				- Trailing zeroes in an address can be omitted when not used
					® Often the last few digits are omitted due to large number of possible combinations
	- Differences
		- Slight oversimplification
			§ IP supports software
			§ MAC supports hardware implementation of the network stack
		- MAC generally remains fixed across networks
		- IP changes as device is moved from network to network
		- Sending a packet into the world
			§ Local network uses MAC between devices
			§ Big internet uses IP between cities/countries/whatever
			§ Goes to the destination local network which uses MAC again
- Ports
	- Physical or virtual connection points
		- Physical
			§ Allow cable connections
		- Virtual
			§ Allow virtual connections
	- Numbering
		- Well known (0-1023)
			§ 20 & 21: FTP
			§ 22: SSH
			§ 23: Telnet
			§ 25: SMTP
			§ 80: HTTP
			§ 443: HTTPS
		- Registered (1024-49151)
		- Dynamic or private (49152-65535)
		- To hide/secure a web server, use a nonstandard port so only you will know it's location

    - Nmap
        - a limited scan is ok
        - a more involved scan can be illegal
        - network scanner  
        - functions
            - ping sweeps
                - `nmap -sP target_ip`
            - TCP connect scans
                - three-way handshake
                    1. send a TCP SYN segment to a target port
                    2. if a SYN/ACK is received
                       - the port is open
                      - respond with an ACK
                     3. if no response, an RST or an ICMP error is received
                - `nmap -sT target_ip`
            - UDP scans
                - send a UDP packet
                    - send a UDP packet to a port
                    - if ICMP port unreachable, the port is closed
                    - if no response, the port is open
                - `nmap -sU target_ip`
            - TCP FIN
                - send a FIN packet
            - Xmas Tree
                - send a packet with all flags set
                    - sends a TCP segment to a port with FIN, URG, PSH flags set
                        - violates TCP RFC
                    - if RST is received, the port is closed
                    - if no response, the port may be open
            - Null
                - send a packet with no flags set
                    - sends a TCP segment to a port with no flags set
                        - violates TCP RFC
                    - if RST is received, the port is closed
                    - if no response, the port may be open
            - TCP ACK
                - send ACK to port
                    - if RST is received, the packet made it through the filter
                    - or the port is unfiltered
                - if no response, the port is filtered
        - options
            - port ranges

            - decoy packets

- Exploiting Vulnerabilities & Metasploit Overview
    - exploits
        - software, data chunks, sequences that take advantage of a vulnerability
        - the goal is to cause unintended behavior
            - e.g. privilege escalation, DoS, etc
        - local vs remote
            - remote occurs over a network
                - does not require any prior access
                - e.g.
                    - old Apache HTTP Server allowed remote user to cause DoS by sending a specially crafted request
            - local
                - requires prior access
                - usually involves increasing privileges beyond what is intended
                - e.g. in old Ubuntu versions, you could overflow the buffer of the "login" portion of the ftp terminal command
                    - fpt client runs at the same permissions as the user so you aren't really gaining anything
            - it is not about physical location
                - e.g. gaining super user access over SSH is considered a local exploit due to the prior access
        - exploit databases
            - websites that list known vulnerabilities
            - e.g. exploit-db.com
            - can be used to find vulnerabilities in a system
    - Metasploit
        - framework for developing and executing exploits
        - modular design
            - exploits
                - code that takes advantage of a vulnerability
            - payloads
                - ccode to run remotely once the exploit gets you in
            - auxiliary
                - tools to gather information
            - encoders
                - tools to convert from one format to another
            - and more...
        - available in the open-source Framework and Pro versions
            - Pro is **really** expensive
            - Framework is free
        - cheat sheet: https://www.ittc.ku.edu/~alexbardas/eecs465/materials/various/metasploit-cheat-sheet.pd
