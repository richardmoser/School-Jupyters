```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
h = 6.626e-34  # Planck's constant [J s]
c = 3.0e8      # Speed of light [m/s]
k_B = 1.38e-23 # Boltzmann's constant [J/K]
pi = np.pi

# Planck function: I_lambda (Intensity as a function of wavelength and temperature)
def planck(wavelength, T):
    return (2 * h * c**2) / (wavelength**5) * (1 / (np.exp((h * c) / (wavelength * k_B * T)) - 1))

# Wavelength range (in meters)
wavelengths = np.linspace(1e-7, 3e-6, 500)  # from 0.1 to 3 micrometers

# Temperatures for the Earth and the Sun
T_earth = 300     # Earth's temperature [K]
T_sun = 5800      # Sun's temperature [K]

# Calculate the flux density for both the Earth and the Sun (F_lambda = pi * I_lambda)
F_lambda_earth = pi * planck(wavelengths, T_earth)
F_lambda_sun = pi * planck(wavelengths, T_sun)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(wavelengths * 1e6, F_lambda_earth, label=f'Earth (T={T_earth}K)', color='blue')  # Wavelength in micrometers
plt.plot(wavelengths * 1e6, F_lambda_sun, label=f'Sun (T={T_sun}K)', color='orange')    # Wavelength in micrometers

# Labeling the plot
plt.title('Flux Density $F_\\lambda$ at the Surface of an Earth-sized Planet and Sun-like Star')
plt.xlabel('Wavelength (micrometers)')
plt.ylabel('Flux Density $F_\\lambda$ (W m$^{-2}$ μm$^{-1}$)')
plt.yscale('log')  # Logarithmic scale to show differences
plt.legend()
plt.grid(True)

# Show plot
# plt.show()

# save to C:\Users\rj\IdeaProjects as PS2_3a.png
plt.savefig('C:\\Users\\rj\\IdeaProjects\\untitled1\\PS2_3a.png')
```

For a neutral H atom, draw an energy diagram for transitions between all energy levels $n \leq 6$. [5 pts]

```python
import matplotlib.pyplot as plt

# Energy levels for hydrogen atom
E_n = [-13.6, -3.4, -1.51, -0.85, -0.54, -0.38]  # Energy
n_values = [1, 2, 3, 4, 5, 6]  # Principal
colors = ['red', 'green', 'blue', 'orange', 'purple', 'brown']  # Colors for
# Plotting
plt.figure(figsize=(10, 6))
for i, E in enumerate(E_n):
    plt.hlines(E, 0, 1, color=colors[i], label=f'n = {n_values[i]}')
    plt.text(1.01, E, f'n = {n_values[i]}', va='center', ha='left', color=colors[i])
plt.ylim(-15, 0)

# Labeling the plot
plt.title('Energy Levels of Hydrogen Atom')
plt.xlabel('Energy (eV)')
plt.ylabel('Energy Level')
plt.legend()
plt.grid(True)

# Show plot
# plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Constants
E0 = 13.6  # Ground state energy in eV

# Calculate energy levels
n_levels = np.arange(1, 7)  # n = 1 to 6
energy_levels = -E0 / n_levels**2  # Energies in eV

# Plot energy levels
plt.figure(figsize=(8, 6))

for n, E in zip(n_levels, energy_levels):
    plt.hlines(E, 0, 1, label=f'n={n}', color='black')

# Label the energy levels
for n, E in zip(n_levels, energy_levels):
    plt.text(1.02, E, f'n={n}', verticalalignment='center')

plt.ylabel('Energy (eV)')
plt.xlabel('Energy Levels')
plt.xlim(0, 1.5)
plt.ylim(-13.6, 0)
plt.title('Energy Levels of Hydrogen Atom (n ≤ 6)')
plt.grid(True)

# Define function to plot transitions
def plot_transition(n_initial, n_final, color):
    E_initial = -E0 / n_initial**2
    E_final = -E0 / n_final**2
    plt.arrow(1, E_initial, 0, E_final - E_initial, head_width=0.05, head_length=0.3, length_includes_head=True, color=color)

# Plot transitions for each series
# Lyman (n -> 1)
for n in range(2, 7):
    plot_transition(n, 1, 'blue')

# Balmer (n -> 2)
for n in range(3, 7):
    plot_transition(n, 2, 'green')

# Paschen (n -> 3)
for n in range(4, 7):
    plot_transition(n, 3, 'red')

# Legend and display
plt.legend()
plt.show()
```

For these same hydrogen transitions (n ≤ 6) plot on a graph or number line, and label, the locations of all
possible photon wavelengths. Use a logarithmic wavelength axis. [5 pts]

```python
import matplotlib.pyplot as plt
import numpy as np

# Constants
Rydberg = 1.097e7  # Rydberg constant in m^-1
    
# Calculate the wavelengths for transitions from n to m
def wavelength(n, m):
    return 1 / (Rydberg * (1/m**2 - 1/n**2))

plt.figure(figsize=(10, 2.5))

# Wavelengths for transitions from n to m (n ≤ 6)
wavelengths = []
for n in range(1, 7):
    for m in range(n+1, 7):
        wavelength_i = abs(wavelength(n, m))
        plt.scatter(wavelength_i, 0, color='red', marker='|', s=3000)
        # plt.text(wavelength_i, 0.01, f'n={n}, m={m}', rotation=90, verticalalignment='bottom')
        wavelengths.append(wavelength_i)
# Plotting
# plt.scatter(wavelengths, np.zeros(len(wavelengths)), color='red', marker='|', s=1000)
plt.yticks([])  # Remove y-axis ticks
plt.xscale('log')  # Logarithmic scale for wavelengths

# Labeling the plot
plt.title('Wavelengths of Hydrogen Transitions (n ≤ 6)')
plt.xlabel('Wavelength (m)')
# plt.grid(True)

# Show plot
# plt.show()

# save to C:\Users\rj\IdeaProjects as PS4_2b.png
plt.savefig('C:\\Users\\rj\\IdeaProjects\\untitled1\\PS4_2b.png')
```
